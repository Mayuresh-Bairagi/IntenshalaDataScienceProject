# Model

```python
traget = train['subscribed']
train = train.drop('subscribed', axis = 1)
```

```python
train = pd.get_dummies(train)
```

```python
train.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_val, y_train, y_val = train_test_split(train, traget, test_size = 0.2, random_state=12)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_model  = LogisticRegression()
```

```python
logistic_model.fit(X_train,y_train)
```

```python
prediction = logistic_model.predict(X_val)
```

```python
prediction
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_val, prediction)
```

### We got an accuracy score 88 % which arounf 90% 

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
Decision_model = DecisionTreeClassifier(max_depth=5, random_state=0)
```

```python
Decision_model.fit(X_train,y_train)
```

```python
prediction_decisionTree = Decision_model.predict(X_val)
```

```python
prediction_decisionTree
```

```python
accuracy_score(y_val, prediction_decisionTree)
```

### We got an accuracy of more than 90% on the validation set

```python
test = pd.get_dummies(test)
test
```

```python
test_prediction = Decision_model.predict(test)
```

```python
submission = pd.DataFrame()
```

```python
submission['ID'] = test['ID']
submission['subscribed'] = test_prediction
```

```python
submission['subscribed'].replace(0,'no',inplace=True)
submission['subscribed'].replace(1,'yes',inplace=True)
```

```python
submission.to_csv('submission.csv', header=True, index=False)
```
